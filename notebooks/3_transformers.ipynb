# RuBERT (DeepPavlov)

```python
import pandas as pd
df_balanced = pd.read_csv('lenta-ru-news_balanced.csv')  # укажи правильный путь к файлу
```

```python
# --- Импорт библиотек ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score
from scipy.special import softmax
from transformers import EarlyStoppingCallback

# --- 1️⃣ Подготовка данных ---

texts = df_balanced['text']
labels = df_balanced['topic']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)

# Кодируем метки
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

num_classes = len(le.classes_)

# --- 2️⃣ Токенизация для BERT ---

model_name = "DeepPavlov/rubert-base-cased"

tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=256)

# Подготовка данных для Hugging Face Dataset
train_data = pd.DataFrame({"text": X_train.tolist(), "label": y_train_enc})
test_data = pd.DataFrame({"text": X_test.tolist(), "label": y_test_enc})

train_ds = Dataset.from_pandas(train_data)
test_ds = Dataset.from_pandas(test_data)

train_ds = train_ds.map(tokenize, batched=True)
test_ds = test_ds.map(tokenize, batched=True)

# Убираем колонки с текстом
train_ds = train_ds.remove_columns(["text"])
test_ds = test_ds.remove_columns(["text"])

train_ds = train_ds.with_format("torch")
test_ds = test_ds.with_format("torch")

# --- 3️⃣ Создание модели ---

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_classes
)

# --- 4️⃣ Метрики для Trainer ---

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)

    accuracy = accuracy_score(labels, preds)
    macro_f1 = f1_score(labels, preds, average='macro')
    weighted_f1 = f1_score(labels, preds, average='weighted')

    return {
        'accuracy': accuracy,
        'macro_f1': macro_f1,
        'weighted_f1': weighted_f1
    }

# --- 5️⃣ Параметры обучения ---

training_args = TrainingArguments(
    output_dir="./rubert_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=64,
    per_device_eval_batch_size=64,
    num_train_epochs=5,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1"
)

# --- 6️⃣ Тренировка модели ---

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
)

trainer.train()

# --- 6.1️⃣ Сохранение модели и Tokenizer ---

trainer.save_model("./rubert_final_model")
tokenizer.save_pretrained("./rubert_final_model")

# Сохраняем LabelEncoder
import pickle
with open("rubert_label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

print("Модель, Tokenizer и LabelEncoder сохранены успешно!")

# --- 7️⃣ Предсказания и метрики ---

predictions = trainer.predict(test_ds)
y_pred_enc = np.argmax(predictions.predictions, axis=1)
y_pred = le.inverse_transform(y_pred_enc)
y_true = y_test.tolist()

print("Classification Report:")
print(classification_report(y_true, y_pred))

accuracy = accuracy_score(y_true, y_pred)
macro_f1 = f1_score(y_true, y_pred, average='macro')
weighted_f1 = f1_score(y_true, y_pred, average='weighted')

# ROC AUC

y_test_bin = label_binarize(y_test_enc, classes=np.arange(len(le.classes_)))
y_pred_probs = softmax(predictions.predictions, axis=1)
roc_auc = roc_auc_score(y_test_bin, y_pred_probs, average='macro')

print(f"\nОсновные метрики:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Macro F1-score: {macro_f1:.4f}")
print(f"Weighted F1-score: {weighted_f1:.4f}")
print(f"ROC AUC (macro): {roc_auc:.4f}")

# --- 8️⃣ Матрица ошибок ---

cm = confusion_matrix(y_true, y_pred, labels=le.classes_)

plt.figure(figsize=(14, 12))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=le.classes_,
    yticklabels=le.classes_
)
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.title('Матрица ошибок (RuBERT)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# инференс rubert + lime keywords
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import pickle
import numpy as np
from scipy.special import softmax
from lime.lime_text import LimeTextExplainer

# --- Загрузка модели и токенизатора ---
model = AutoModelForSequenceClassification.from_pretrained("./rubert_final_model")
tokenizer = AutoTokenizer.from_pretrained("./rubert_final_model")

# --- Загрузка LabelEncoder ---
with open("rubert_label_encoder.pkl", "rb") as f:
    le = pickle.load(f)

# --- Список категорий (из LabelEncoder) ---
categories = list(le.classes_)

# --- Функция предсказания для LIME ---
def predict_proba(texts):
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True, max_length=256)
    with torch.no_grad():
        outputs = model(**inputs)
    probs = softmax(outputs.logits.numpy(), axis=1)
    return probs

# --- Основная функция инференса + объяснение ---
def predict_and_explain(text, num_features=8):
    # Предсказание
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=256)
    with torch.no_grad():
        outputs = model(**inputs)

    logits = outputs.logits.detach().numpy()
    probs = softmax(logits, axis=1)
    pred_class_idx = np.argmax(probs)
    class_name = le.inverse_transform([pred_class_idx])[0]
    confidence = probs[0][pred_class_idx]

    print(f"🧠 Предсказанная категория: {class_name}")
    print(f"📊 Уверенность модели: {confidence:.2%}")

    # Объяснение с LIME
    explainer = LimeTextExplainer(class_names=categories)
    explanation = explainer.explain_instance(text, predict_proba, num_features=num_features, top_labels=1, num_samples=50)

    print("🔍 Ключевые слова по мнению модели:")
    for word, weight in explanation.as_list(label=pred_class_idx):
        print(f"— {word}: {round(weight, 4)}")

# --- Пример использования ---
text_example = "После нескольких месяцев споров и взаимных угроз европейские страны, обсуждавшие возможность отправки «миротворцев» на Украину"
predict_and_explain(text_example)
```

```python
# keywords from rubert
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import numpy as np

# --- Загрузка модели ---
model_name = "./rubert_final_model"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
model.eval()

def get_word_importance(text):
    # Токенизация с оффсетами
    inputs = tokenizer(text, return_tensors="pt", return_offsets_mapping=True, truncation=True, padding=True, max_length=256)
    
    # Получение эмбеддингов и включение автодифференцировки
    embedding_layer = model.get_input_embeddings()
    embedded = embedding_layer(inputs['input_ids'])
    embedded.requires_grad_()
    embedded.retain_grad()

    # Предсказание по embedded
    outputs = model(inputs_embeds=embedded, attention_mask=inputs['attention_mask'])
    pred_class = torch.argmax(outputs.logits, dim=1)
    outputs.logits[0, pred_class].backward()

    # Градиенты и веса
    grads = embedded.grad[0]  # (seq_len, hidden_dim)
    grads_norm = grads.norm(dim=1)  # (seq_len,)

    # --- Преобразование токенов обратно в слова ---
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])
    offsets = inputs['offset_mapping'][0].tolist()

    word_scores = {}
    current_word = ""
    current_score = 0
    current_len = 0

    for token, offset, score in zip(tokens, offsets, grads_norm):
        if token in ["[CLS]", "[SEP]"]:
            continue
        if token.startswith("##"):
            current_word += token[2:]
            current_score += score.item()
            current_len += 1
        else:
            if current_word:
                word_scores[current_word] = current_score / max(current_len, 1)
            current_word = token
            current_score = score.item()
            current_len = 1

    # Добавить последний токен
    if current_word:
        word_scores[current_word] = current_score / max(current_len, 1)

    # Сортировка по важности
    sorted_words = sorted(word_scores.items(), key=lambda x: x[1], reverse=True)

    return sorted_words[:5]


# --- Пример ---
text = "Экспортные цены на российскую пшеницу продолжили падать. Об этом со ссылкой на аналитиков Русагротранса сообщает «Интерфакс».  За последнюю неделю стоимость зерна с поставкой из РФ упала с 248 долларов за тонну до 225-229 долларов. В то же время стоимость французской пшеницы снизилась до 229 долларов, румынской — до 246 долларов, аргентинской — до 233 долларов, австралийской — до 247 долларов. Американское зерно подорожало сразу на семь долларов, до 226 за тонну. Зерно дорожало на фоне засухи на юге России, жары в Китае и ухудшения состояния посевов в США. Однако после ситуация изменилась.  «Дожди в Центральной России и в большей части Поволжья будут поддерживать благоприятные условия для развития посевов», — пояснили в Русагротрансе. В середине мая стоимость фьючерсных контрактов на июльские поставки пшеницы в моменте упали до девятимесячного минимума и достигли отметки около 190 долларов за тонну. Котировки падали до уровня 189,23 доллара за тонну впервые с июля 2024 года. Максимальный показатель в 2025-м был достигнут в феврале. В сравнении с ним майский результат просел более чем на 17 процентов. "
#predict_new_text(text)
important_words = get_word_importance(text)

print("🔍 Важные слова по модели RuBERT (по градиентам):")
for word, score in important_words:
    print(f"— {word}: {round(score, 4)}")
```

 # RuRoBERTa 

```python
import pandas as pd
df_balanced = pd.read_csv('lenta-ru-news_balanced.csv')  # укажи правильный путь к файлу
```

```python
# --- Импорт библиотек ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

# --- 1️⃣ Подготовка данных ---

texts = df_balanced['text']
labels = df_balanced['topic']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)

# Кодируем метки
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

num_classes = len(le.classes_)

# --- 2️⃣ Токенизация для RuRoBERTa ---

model_name = "sberbank-ai/ruRoberta-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=256)

# Подготовка данных для Hugging Face Dataset
train_data = pd.DataFrame({"text": X_train.tolist(), "label": y_train_enc})
test_data = pd.DataFrame({"text": X_test.tolist(), "label": y_test_enc})

train_ds = Dataset.from_pandas(train_data)
test_ds = Dataset.from_pandas(test_data)

train_ds = train_ds.map(tokenize, batched=True)
test_ds = test_ds.map(tokenize, batched=True)

train_ds = train_ds.remove_columns(["text"])
test_ds = test_ds.remove_columns(["text"])

train_ds = train_ds.with_format("torch")
test_ds = test_ds.with_format("torch")

# --- 3️⃣ Создание модели ---

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_classes
)

# --- 4️⃣ Метрики для Trainer ---

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)

    accuracy = accuracy_score(labels, preds)
    macro_f1 = f1_score(labels, preds, average='macro')
    weighted_f1 = f1_score(labels, preds, average='weighted')

    return {
        'accuracy': accuracy,
        'macro_f1': macro_f1,
        'weighted_f1': weighted_f1
    }

# --- 5️⃣ Параметры обучения ---

training_args = TrainingArguments(
    output_dir="./ruroberta_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1"
)

# --- 6️⃣ Тренировка модели ---

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    compute_metrics=compute_metrics
)

trainer.train()

# --- 6.1️⃣ Сохранение модели и Tokenizer ---

trainer.save_model("./ruroberta_final_model")
tokenizer.save_pretrained("./ruroberta_final_model")

# Сохраняем LabelEncoder
import pickle
with open("ruroberta_label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

print("Модель, Tokenizer и LabelEncoder успешно сохранены!")

# --- 7️⃣ Предсказания и метрики ---
predictions = trainer.predict(test_ds)
y_pred_enc = np.argmax(predictions.predictions, axis=1)
y_pred = le.inverse_transform(y_pred_enc)
y_true = y_test.tolist()

# ROC AUC (macro)
y_test_bin = label_binarize(y_test_enc, classes=np.arange(len(le.classes_)))
y_pred_prob = torch.nn.functional.softmax(torch.tensor(predictions.predictions), dim=1).numpy()
roc_auc = roc_auc_score(y_test_bin, y_pred_prob, average='macro')

# Классические метрики
print("Classification Report:")
print(classification_report(y_true, y_pred))

accuracy = accuracy_score(y_true, y_pred)
macro_f1 = f1_score(y_true, y_pred, average='macro')
weighted_f1 = f1_score(y_true, y_pred, average='weighted')

print(f"\nОсновные метрики:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Macro F1-score: {macro_f1:.4f}")
print(f"Weighted F1-score: {weighted_f1:.4f}")
print(f"ROC AUC (macro): {roc_auc:.4f}")

# --- 8️⃣ Матрица ошибок ---

cm = confusion_matrix(y_true, y_pred, labels=le.classes_)

plt.figure(figsize=(14, 12))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=le.classes_,
    yticklabels=le.classes_
)
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.title('Матрица ошибок (RuRoBERTa)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
df_balanced = pd.read_csv('lenta-ru-news_balanced.csv')  # укажи правильный путь к файлу
```

```python
# --- Импорт библиотек ---
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score

from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from datasets import Dataset
import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score

# --- 1️⃣ Подготовка данных ---

texts = df_balanced['text']
labels = df_balanced['topic']

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)

# Кодируем метки
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

num_classes = len(le.classes_)

# --- 2️⃣ Токенизация для rubert-tiny ---

model_name = "cointegrated/rubert-tiny"

tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize(batch):
    return tokenizer(batch["text"], padding="max_length", truncation=True, max_length=256)

# Подготовка данных для Hugging Face Dataset
train_data = pd.DataFrame({"text": X_train.tolist(), "label": y_train_enc})
test_data = pd.DataFrame({"text": X_test.tolist(), "label": y_test_enc})

train_ds = Dataset.from_pandas(train_data)
test_ds = Dataset.from_pandas(test_data)

train_ds = train_ds.map(tokenize, batched=True)
test_ds = test_ds.map(tokenize, batched=True)

train_ds = train_ds.remove_columns(["text"])
test_ds = test_ds.remove_columns(["text"])

train_ds = train_ds.with_format("torch")
test_ds = test_ds.with_format("torch")

# --- 3️⃣ Создание модели ---

model = AutoModelForSequenceClassification.from_pretrained(
    model_name,
    num_labels=num_classes
)

# --- 4️⃣ Метрики для Trainer ---

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)

    accuracy = accuracy_score(labels, preds)
    macro_f1 = f1_score(labels, preds, average='macro')
    weighted_f1 = f1_score(labels, preds, average='weighted')

    return {
        'accuracy': accuracy,
        'macro_f1': macro_f1,
        'weighted_f1': weighted_f1
    }

# --- 5️⃣ Параметры обучения ---

training_args = TrainingArguments(
    output_dir="./rubert_tiny_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=512,  # Можно увеличить из-за маленькой модели
    per_device_eval_batch_size=512,
    num_train_epochs=10,
    weight_decay=0.01,
    logging_dir="./logs",
    load_best_model_at_end=True,
    metric_for_best_model="macro_f1"
)

# --- 6️⃣ Тренировка модели ---

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    compute_metrics=compute_metrics
)

trainer.train()

# --- 6.1️⃣ Сохранение модели, Tokenizer и LabelEncoder ---

trainer.save_model("./rubert_tiny_final_model")
tokenizer.save_pretrained("./rubert_tiny_final_model")

import pickle
with open("rubert_tiny_label_encoder.pkl", "wb") as f:
    pickle.dump(le, f)

print("Модель, Tokenizer и LabelEncoder успешно сохранены!")


# --- 7️⃣ Предсказания и метрики ---

# Предсказания
predictions = trainer.predict(test_ds)
y_pred_logits = predictions.predictions
y_pred_enc = np.argmax(y_pred_logits, axis=1)
y_pred = le.inverse_transform(y_pred_enc)
y_true = y_test.tolist()

# Классические метрики
print("Classification Report:")
print(classification_report(y_true, y_pred))

accuracy = accuracy_score(y_true, y_pred)
macro_f1 = f1_score(y_true, y_pred, average='macro')
weighted_f1 = f1_score(y_true, y_pred, average='weighted')

# ROC AUC (macro)
y_test_bin = label_binarize(y_test_enc, classes=np.arange(len(le.classes_)))
y_pred_prob = torch.nn.functional.softmax(torch.tensor(y_pred_logits), dim=1).numpy()
roc_auc = roc_auc_score(y_test_bin, y_pred_prob, average='macro')

print(f"\nОсновные метрики:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Macro F1-score: {macro_f1:.4f}")
print(f"Weighted F1-score: {weighted_f1:.4f}")
print(f"ROC AUC (macro): {roc_auc:.4f}")

# --- 8️⃣ Матрица ошибок ---

cm = confusion_matrix(y_true, y_pred, labels=le.classes_)

plt.figure(figsize=(14, 12))
sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=le.classes_,
    yticklabels=le.classes_
)
plt.ylabel('Истинный класс')
plt.xlabel('Предсказанный класс')
plt.title('Матрица ошибок (RuBERT-tiny)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import json

# --- 📥 Загрузка логов ---
with open("rubert_tiny_log_history.json", "r", encoding='utf-8') as f:
    logs = json.load(f)

log_df = pd.DataFrame(logs)
log_df = log_df[log_df["epoch"].notna()]  # удалим строки без эпохи (промежуточные шаги)

# --- 📊 Подготовка агрегированных данных по эпохам ---
epoch_df = log_df.groupby("epoch").mean(numeric_only=True).reset_index()
epoch_df = epoch_df.interpolate(method="linear")

# --- 📈 Loss ---
if {"loss", "eval_loss"}.issubset(epoch_df.columns):
    plt.figure(figsize=(10, 5))
    plt.plot(epoch_df["epoch"], epoch_df["loss"], label="Train Loss", linewidth=2)
    plt.plot(epoch_df["epoch"], epoch_df["eval_loss"], label="Validation Loss", linewidth=2)
    plt.xlabel("Эпоха")
    plt.ylabel("Loss")
    plt.title("Потери модели (Loss)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 📈 Accuracy ---
if "eval_accuracy" in epoch_df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(epoch_df["epoch"], epoch_df["eval_accuracy"], label="Validation Accuracy", color='blue', linewidth=2)
    plt.xlabel("Эпоха")
    plt.ylabel("Accuracy")
    plt.title("Точность на валидации")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 📈 Macro F1 ---
if "eval_macro_f1" in epoch_df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(epoch_df["epoch"], epoch_df["eval_macro_f1"], label="Validation Macro F1", color='green', linewidth=2)
    plt.xlabel("Эпоха")
    plt.ylabel("Macro F1")
    plt.title("Macro F1 на валидации")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 📈 Weighted F1 ---
if "eval_weighted_f1" in epoch_df.columns:
    plt.figure(figsize=(10, 5))
    plt.plot(epoch_df["epoch"], epoch_df["eval_weighted_f1"], label="Validation Weighted F1", color='purple', linewidth=2)
    plt.xlabel("Эпоха")
    plt.ylabel("Weighted F1")
    plt.title("Weighted F1 на валидации")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 📈 Learning Rate ---
if "learning_rate" in log_df.columns:
    lr_df = log_df[["step", "learning_rate"]].dropna().sort_values("step")
    if not lr_df.empty:
        lr_df = lr_df.set_index("step").interpolate().reset_index()
        plt.figure(figsize=(10, 5))
        plt.plot(lr_df["step"], lr_df["learning_rate"], label="Learning Rate", color="blue", linewidth=2)
        plt.xlabel("Шаг")
        plt.ylabel("Learning Rate")
        plt.title("Изменение Learning Rate")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

# --- 📈 Gradient Norm ---
if "grad_norm" in log_df.columns:
    gn_df = log_df[["step", "grad_norm"]].dropna().sort_values("step")
    if not gn_df.empty:
        gn_df = gn_df.set_index("step").interpolate().reset_index()
        plt.figure(figsize=(10, 5))
        plt.plot(gn_df["step"], gn_df["grad_norm"], label="Gradient Norm", color="red", linewidth=2)
        plt.xlabel("Шаг")
        plt.ylabel("Grad Norm")
        plt.title("Норма градиента на каждом шаге")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

# --- 📈 Eval Samples/sec ---
if "eval_samples_per_second" in log_df.columns and log_df["eval_samples_per_second"].count() > 1:
    eval_perf_df = log_df[["epoch", "eval_samples_per_second"]].dropna().sort_values("epoch")
    eval_perf_df = eval_perf_df.set_index("epoch").interpolate().reset_index()

    plt.figure(figsize=(10, 5))
    plt.plot(eval_perf_df["epoch"], eval_perf_df["eval_samples_per_second"],
             label="Eval Samples/sec", color="darkblue", linewidth=2)
    plt.xlabel("Эпоха")
    plt.ylabel("Samples/sec")
    plt.title("Скорость обработки на валидации (Eval Samples/sec)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# --- 📈 Train Samples/sec ---
if "train_samples_per_second" in log_df.columns and log_df["train_samples_per_second"].count() > 1:
    train_perf_df = log_df[["step", "train_samples_per_second"]].dropna().sort_values("step")
    train_perf_df = train_perf_df.set_index("step").interpolate().reset_index()

    plt.figure(figsize=(10, 5))
    plt.plot(train_perf_df["step"], train_perf_df["train_samples_per_second"],
             label="Train Samples/sec", color="brown", linewidth=2)
    plt.xlabel("Шаг")
    plt.ylabel("Samples/sec")
    plt.title("Скорость обучения (Train Samples/sec)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import pickle
import numpy as np
from scipy.special import softmax

# --- Загрузка модели и токенизатора ---
model = AutoModelForSequenceClassification.from_pretrained("./rubert_tiny_final_model")
tokenizer = AutoTokenizer.from_pretrained("./rubert_tiny_final_model")

with open("rubert_tiny_label_encoder.pkl", "rb") as f:
    le = pickle.load(f)

# --- Функция предсказания с уверенностью ---
def predict_new_text(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=256)
    with torch.no_grad():
        outputs = model(**inputs)

    logits = outputs.logits.detach().numpy()
    probs = softmax(logits, axis=1)
    pred_class_idx = np.argmax(probs)
    class_name = le.inverse_transform([pred_class_idx])[0]
    confidence = probs[0][pred_class_idx]

    print(f"Предсказанная категория: {class_name}")
    print(f"Уверенность модели: {confidence:.4f}")

# --- Пример использования ---
predict_new_text("После нескольких месяцев споров и взаимных угроз европейские страны, обсуждавшие возможность отправки «миротворцев» на Украину, похоже, окончательно отказались от этой идеи. Причина — страх перед возможной новой мировой войной или как минимум неконтролируемой эскалацией в противостоянии с Россией. Переосмыслили свою позицию и США: как пишет The Telegraph, представители команды Дональда Трампа в частном порядке уведомили союзников о возможности предоставить «коалиции желающих» гарантии безопасности для интервенции — уже после заключения мира. О значении этого шага в контексте переговоров и возможной реакции России — в материале «Ленты.ру».")
```
